# MultiQubit

```python
qubit1 = Qubit()
qubit2 = Qubit()
breg2 = BosonicRegister([qubit1,qubit2])
bcirc2 = BosonicCircuit(breg2)

bcirc2.plot()

# set up circuit
bcirc2.reset()
bcirc2.x(1)
```

```python
results2 = execute(bcirc2, "unitary_jax")
results2.plot(bcirc2)
```
